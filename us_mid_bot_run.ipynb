```python
# get UK dataframe
# what changes when country changes -- cd and dflist name
cd = '/Users/melia/Downloads/country_query_results/UK_nocsv'

dfListUK = []
frame0 = pd.DataFrame()

for root, dirs, files in os.walk(cd):
    for fname in files:
        if re.match("^.*.csv$", fname):
            frame = pd.read_csv(os.path.join(root, fname))
            #frame['key'] = "file{}".format(i)
            frame0 = pd.concat([frame0, frame], axis=1)
            dfListUK.append(frame0)
```

```python
# what changes when country changes -- cd, dflist name, df name, local

df_UK = dfListUK[-1] # still need to change this too
print(df_UK.shape)

df_UK.columns = list(range(1,df_UK.shape[1]+1))

for i in range(1, df_UK.shape[1]+1):
    df_UK[f'parsed{i}'] = pd.to_datetime(df_UK[i], format='%Y-%m-%d %H:%M:%S')
    df_UK[f'minute_of_week{i}'] = df_UK[f'parsed{i}'].apply(lambda row: 
                                            (row.dayofweek * 24 * 60) + (row.hour * 60) + (row.minute))
    
```

```python
df_us_uk = df_us_uk.drop(columns=['index'])
df_us_uk.head()
```

```python
# add local back - definitely a better way to do this
df_us_uk['local2'] = np.where(df_us_uk.index <= 71, 1, 2)
```

```python
# shuffle rows in dataframe
df_us_uk_shuf = df_us_uk.sample(frac=1).reset_index(drop=True)
df_us_uk_shuf.tail()
```

```python
# confirm values are actually different
df_us_uk_shuf.head()
```

```python
# now it's actually time to model
# set X and y -- make sure no NaN
X = df_us_uk_shuf.iloc[:, :1500]
y = df_us_uk_shuf['local2']
```

```python

# for knn usually you need to do some preprocessing but here we have all the same units for features 
knn = KNeighborsClassifier(n_neighbors=5, weights='distance', n_jobs=-1)
knn.fit(X, y)
```

```python
# cross val score for knn
print(cross_val_score(knn, X, y, cv=10))
```

```python
rfc = RandomForestClassifier(n_estimators=100, n_jobs=-1)
rfc.fit(X, y)
# can make training and testing sets too
```

```python
# rfc scores
print(cross_val_score(rfc, X, y, cv=10))
```
